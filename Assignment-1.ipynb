```python
# using the PIVOT technique to categorize cities based on similar weather

df1=df.pivot(columns="Weather",values="City")
df1.fillna('-',inplace=True) # filling Nan to -
# saving the obtained data drame to df1
df1.to_csv('C:\\Users\\DELL\\Desktop\\MY STUFF\DATA SCIENCE\DATA\\Raw data\\similar_weather_cities.csv')
```
